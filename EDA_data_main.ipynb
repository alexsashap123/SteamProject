## **Piechart for genre distribution**

```python
gf = df.copy()
gf['all_genres'] = gf['genres'].str.split(', ')
gf = gf.explode('all_genres')

all_genres_count = gf['all_genres'].value_counts()

stop = 2.8
main_genres = all_genres_count[all_genres_count / all_genres_count.sum() * 100 >= stop]
other_genres = all_genres_count[all_genres_count / all_genres_count.sum() * 100 < stop]

combined = main_genres.copy()
combined['Others'] = other_genres.sum()

colors = sns.color_palette("pastel", len(combined) - 1)
colors.append('Lightsalmon')

fig, ax = plt.subplots(figsize=(8, 8))
plt.pie(combined.values, colors=colors, autopct='%1.1f%%', startangle=90)

plt.title(f'Топ 10 популярных жанров',
         fontfamily='monospace',
         fontsize=16,
         fontweight='bold')

ax.legend(combined.index, loc='center left', bbox_to_anchor=(1, 0.5),
          title="Genres", prop={'family': 'monospace'})

plt.setp(ax.texts, fontfamily='monospace', fontsize=12)
plt.tight_layout()

plt.show()
```

## **Visualisation for Average user rating by release year**

```python
df['released_year'] = pd.to_datetime(df['released']).dt.year

year_df = df[df['rating'].notna()].groupby('released_year')['rating'].mean().reset_index()

plt.figure(figsize=(16, 8))

plt.plot(year_df['released_year'], year_df['rating'],
         marker='o',
         linewidth=2,
         markersize=6,
         markerfacecolor='red',
         markeredgecolor='darkred',
         markeredgewidth=1,
         color='steelblue')

for i, row in year_df.iterrows():
    plt.annotate(f"{row['rating']:2f}",
                (row['released_year'], row['rating']),
                textcoords="offset points",
                xytext=(0,10),
                ha='center',
                fontsize=9,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))

plt.title('Average game rating by release year', fontsize=16, fontweight='bold')

plt.xticks(year_df['released_year'], rotation=45, ha='right')

min_rating = year_df['rating'].min()
max_rating = year_df['rating'].max()
plt.yticks(np.arange(round(min_rating, 1), round(max_rating, 1) + 0.1, 0.1))

plt.xlabel('Release Year', fontsize=12, fontweight='bold')
plt.ylabel('Average Rating', fontsize=12, fontweight='bold')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## **Visualisation for Average Metacritic Score by Release Year**

```python
df['released_year'] = pd.to_datetime(df['released']).dt.year

year_df = df[df['metacritic_final'].notna()].groupby('released_year')['metacritic_final'].mean().reset_index()

plt.figure(figsize=(16, 8))

plt.plot(year_df['released_year'], year_df['metacritic_final'],
         marker='o',
         linewidth=2,
         markersize=6,
         markerfacecolor='red',
         markeredgecolor='darkred',
         markeredgewidth=1,
         color='steelblue')

for i, row in year_df.iterrows():
    plt.annotate(f"{row['metacritic_final']:2f}",
                (row['released_year'], row['metacritic_final']),
                textcoords="offset points",
                xytext=(0,10),
                ha='center',
                fontsize=9,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))

plt.title('Average Metacritic Score by Release Year', fontsize=16, fontweight='bold')

xticks_years = year_df['released_year'][::2]
plt.xticks(xticks_years, rotation=45, ha='right')

min_rating = year_df['metacritic_final'].min()
max_rating = year_df['metacritic_final'].max()
intr = year_df['metacritic_final'].round().astype(int).unique()
plt.yticks(intr)

plt.xlabel('Release Year', fontsize=12, fontweight='bold')
plt.ylabel('Average Metacritic Score', fontsize=12, fontweight='bold')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**ЦЕЛЕВАЯ АУДИТОРИЯ**

Основная целевая аудитория представленных игр это Everyone (0+) и Teen (13+) - около 60% от всех возрастных категорий. Это означает, что игры в основном расчитаны на широкую аудиторию и подходят всем возрастным категориям, а также расчитаны на специфичную подростковую аудиторию, которая является одной из самых активных.

При этом самой активной аудиторией является Mature и Teen. Именно они больше всего взаимодействуют с играми и дают обратную связь.

Для аудитории Teen: Action, Adventure, RPG, Indie
