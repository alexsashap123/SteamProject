```python
import pandas as pd
import numpy as np
import glob
files = glob.glob('final_result_*.csv')
tables = []
for file in sorted(files):
    try:
        df = pd.read_csv(file)
        tables.append(df)
        print(f"Успешно загружен: {file}")
    except Exception as e:
        print(f"Ошибка при загрузке {file}: {e}")
if tables:
    df = pd.concat(tables, ignore_index=True)
    print(f"Конкатенация завершена! Итоговый размер: {df.shape}")
else:
    print("Не удалось загрузить ни одного файла")
df.to_csv('final_result.csv', index = False)
```

```python
import pandas as pd
import re
```

```python
data = pd.read_csv('final_result.csv')
```

```python

data['platforms'] = data['platforms'].fillna('')
data['platforms'] = data['platforms'].apply(lambda x: re.findall(r"'name':\s*'([^']*)'", x))
```

```python

data['esrb_rating'] = data['esrb_rating'].fillna('')
data['esrb_rating'] = data['esrb_rating'].apply(lambda x: re.findall(r"'name':\s*'([^']*)'", x))
```

```python

data['genres'] = data['genres'].fillna('')
data['genres'] = data['genres'].apply(lambda x: re.findall(r"'name':\s*'([^']*)'", x))
```

```python

data['stores'] = data['stores'].fillna('')
data['stores'] = data['stores'].apply(lambda x: re.findall(r"'name':\s*'([^']*)'", x))
```

```python

ratingmeta = {
    'Rated M': 'Mature',
    'Rated T': 'Teen',
    'Rated E': 'Everyone',
    'Rated E +10': 'Everyone 10+',
    'Rated AO': 'Adults Only',
    'Rated RP': 'Rating Pending'
}

data['metacritic_rating'] = data['metacritic_rating'].replace(ratingmeta)

data['metacritic_rating'] = data['metacritic_rating'].where(
    data['metacritic_rating'].isin(ratingmeta.values()),
    'No rating'
)
```

```python
data['platforms'] = data['platforms'].apply(lambda lst: ', '.join(lst))
```

```python
data['stores'] = data['stores'].apply(lambda lst: ', '.join(lst))
```

```python
data['genres'] = data['genres'].apply(lambda lst: ', '.join(lst))
```

```python
data['esrb_rating'] = data['esrb_rating'].apply(lambda lst: ', '.join(lst))
```

```python
data['age_rating'] = data['metacritic_rating'].where(data['metacritic_rating'] != 'No rating', data['esrb_rating'])
```

```python
data['metacritic_final'] = data['metacritic_score'].where(data['metacritic_score'].notna(), data['metacritic'])
```
